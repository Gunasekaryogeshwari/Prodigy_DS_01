IMPORTING NECESSARY LIBRARIES

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

UPLOADING THE REQUIRED DATASET FOR THE OPERATION

```python
data = pd.read_csv("/content/train.csv")
```

TO CHECK FOR MISSING VALUES

```python
data.isnull().sum()
```

HANDLE OF THE NULL VALUES

```python
data['Embarked'] = data['Embarked'].fillna('s')
data['Age'] = data['Age'].fillna(data['Age'].mean())
data['Cabin'] = data['Cabin'].fillna('Missing')
```

```python
data.isnull().sum()
```

PICTORIAL REPRESENTATION OF THE DATASET

```python
plt.figure(figsize=(8, 4))
sns.histplot(data["Age"], kde=True, color = "green")
plt.title("Distribution of passengers AGE", fontsize = 20,color = "maroon")
plt.xlabel("Age",fontsize = 15 , color = 'maroon')
plt.ylabel("Count" , fontsize = 15 , color = 'maroon')
plt.show()
```

```python
plt.figure(figsize=(8, 4))  # Adjust figure size for horizontal layout
sns.barplot(
    data=data,
    x="Sex",
    y="Survived",
    hue="Survived",
    orient="h",
    color="lightgrey",
    palette="deep",
)
plt.title("Survival Rate by Gender")
plt.xlabel("Survival")

# Replace legend labels with actual values
handles, labels = plt.gca().get_legend_handles_labels()
new_labels = [label.replace("0", "Male").replace("1", "Female") for label in labels]
plt.legend(handles, new_labels, title="Survived", loc="upper right")

# Alternatively, set custom y-axis ticks and labels (excluding origin)
plt.yticks([0.5, 1.5], ["Male", "Female"])

# Choose one option based on your visualization (Matplotlib or Seaborn)

plt.show()
```
